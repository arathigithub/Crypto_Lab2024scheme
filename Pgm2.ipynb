```python
%%writefile bitwise_string.c
#include <stdio.h>

int main() {
    char *str = "Hello world";

    printf("Char\tASCII\tAND\tOR\tXOR\n");
    printf("---------------------------------\n");

    for(int i = 0; str[i] != '\0'; i++) {
        printf("%c\t%d\t%d\t%d\t%d\n",
               str[i],
               str[i],
               str[i] & 127,
               str[i] | 127,
               str[i] ^ 127);
    }

    return 0;
}
```
